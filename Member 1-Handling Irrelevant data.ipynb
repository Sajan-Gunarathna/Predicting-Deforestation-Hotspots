```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# --- Snippet 1: Load Raw Data and Show the Problem ---
print("--- Loading Raw Data to Show All Columns ---")
# We load the full, messy,noisy dataset with all its original columns
df_raw = pd.read_csv('FAOSTAT_data_en_8-15-2025.csv')

print(f"The original dataset has {len(df_raw.columns)} columns.")
print("Original Columns List:")
print(list(df_raw.columns))
```

```python
# --- Snippet 2: Apply the Solution (Select Relevant Columns) ---
print("\n--- Applying the Solution: Selecting Only Relevant Columns ---")
# We define a list of only the columns we need for our analysis LITERALLY NOTHING ELSE
relevant_cols = ['Area', 'Item', 'Year', 'Value']
df_cleaned = df_raw[relevant_cols]

print(f"The cleaned dataset now has {len(df_cleaned.columns)} columns.")
print("Cleaned Columns List:")
print(list(df_cleaned.columns))
```

```python
# Number of columns
total = len(df_raw.columns)
kept = len(df_cleaned.columns)
removed = total - kept

# Labels and values
labels = ['Total', 'Kept', 'Removed']
values = [total, kept, removed]

plt.figure(figsize=(8, 5))
bars = plt.bar(labels, values, color=['#ffcc99', '#66b3ff', '#ff9999'])
plt.ylabel('Number of Columns')
plt.title('Columns Before and After Cleaning')

# Add numbers on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.2, int(yval), ha='center', va='bottom')

plt.show()
```

```python
# Your relevant columns
relevant_cols = ['Area', 'Item', 'Year', 'Value']

# All columns in the raw dataset
all_cols = df_raw.columns.tolist()

# Create a binary status: 1 if kept, 0 if removed
column_status = [1 if col in relevant_cols else 0 for col in all_cols]

# Convert to DataFrame (1 row, all columns)
df_status = pd.DataFrame([column_status], columns=all_cols)
```

```python
plt.figure(figsize=(12, 2))
sns.heatmap(df_status, annot=True, cbar=False, cmap='YlGnBu', linewidths=0.5)
plt.title('Columns Kept (1) vs Removed (0)')
plt.yticks([])  # hide row labels
plt.show()
```

```python
# --- Final Step: Save the Preprocessed Data ---
output_filename = 'member_1_data_columns_cleaned.csv'
df_cleaned.to_csv(output_filename, index=False)
print(f"\n--- Task Complete ---")
print(f"The dataset with irrelevant columns removed has been saved as '{output_filename}'")
```
